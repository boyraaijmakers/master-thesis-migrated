```python
import pandas as pd
import requests
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas_datareader as web
import pandas as pd
import datetime
import math


results_df = pd.read_csv(r"C:\Users\braaijmakers\Desktop\cogcomp.csv", encoding="utf-8", sep=";")
```

```python
def dbpedia_spotlight(company):
    url = "http://localhost:2222/rest/annotate?text='" + str(company) + "'";
    r = requests.get(url = url) 
    return r.json()
    
    
def dbpedia_query(company):
    sparql = SPARQLWrapper("http://dbpedia.org/sparql")
    sparql.setQuery("""
    PREFIX dbp: <http://dbpedia.org/property/>
    PREFIX owl: <http://www.w3.org/2002/07/owl#>
    select ?stockSymbol ?similarEntities where {
      OPTIONAL { <""" + company.strip() + """> dbp:symbol ?stockSymbol. }
    }
    """)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()["results"]["bindings"]

def wikidata_query(entity):
    sparql = SPARQLWrapper("http://query.wikidata.org/sparql")
    sparql.setQuery("""
    SELECT DISTINCT ?ticker
    WHERE {
        wd:""" + entity + """ p:P414 ?exchange . 
        ?exchange pq:P249 ?ticker .
        FILTER NOT EXISTS {
          wd:""" + entity + """ p:P414 ?exchange . 
          ?exchange pq:P249 ?ticker ;
                    pq:P582 ?end.
        }
    }
    """)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()["results"]["bindings"]            
```

```python
for _, row in results_df.iterrows():
    
    company = row["company"]
    
    if company == "":
        print("-")
        continue
        
    spot = dbpedia_spotlight(company)
    if "Resources" in spot.keys():
        results = dbpedia_query(spot["Resources"][0]["@URI"])
    else:
        print("-")
        continue         
    
    found = False
    
    for result in results:
        if "stockSymbol" in result.keys():
            print(result["stockSymbol"]["value"])
            found = True
            break
        if "similar" in result.keys():
            if "wikidata" in result["similar"]["value"]:
                wikires = wikidata_query(result["similar"]["value"].split("/")[-1])
                if wikires != []:
                    print(wikires[0]["ticker"]["value"])
                    found = True
                    break
    if not found:
        print("--")
```
